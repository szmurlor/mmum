```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.svm import SVR

x=np.arange(0,2*np.pi,0.01); 
d=np.sin(x)
x=np.array( [x] ).T; 
svr = SVR(kernel='rbf', gamma='auto', 
          epsilon=0.1, C=1)
svr.fit(x,d)

y = svr.predict(x)
plt.plot(x,d,'r')
plt.plot(x,y, '--b')
plt.xlabel('x')
plt.ylabel('sin(x), aprokzymacja SVM')
plt.legend(['sin(x)','aproksymacja SVM'])
plt.grid()
```

```python
from scipy.io import loadmat
from sklearn.model_selection import train_test_split

papierosy = loadmat('papierosy.mat')
x = papierosy['x']
d = papierosy['d']
xd = np.concatenate( (x,d), axis=1 )
xd

xu,xt,du,dt = train_test_split(x,d[:,0], train_size=1000)

from sklearn.svm import SVC
svc = SVC(C=1, tol=1e-3, degree=3, gamma='auto')
svc.fit(xu,du)

y = svc.predict(xt)
n_errors = np.sum(y!=dt)
print(f"n_errors={n_errors}")
print(f"error = {n_errors / len(dt)*100}%")

plt.plot(y!=dt)
plt.xlabel('Kolejne prbki testujce')
plt.ylabel('Bd rozpoznania klasy')
```

```python
from numpy.random import randn
x1=1+randn(20,5)
x2=-2+randn(20,5)
x3=3+randn(20,5)
x=np.concatenate((x1,x2,x3),axis=0)
d=np.concatenate( ([1]*20,[2]*20,[3]*20) , axis=0)
```

```python
d
```

```python
# Przykład użycia klasy 


B = RandomForestClassifier(50, 
                oob_score=True, 
                max_samples=3)
B.fit(x,d) 

plt.bar(np.arange(0,5), B.feature_importances_)

# Inna metoda oceny ważności cech
# brak odpowiednika

# out-of-bag error
def oobError(B):
    error = []
    for K in range(1,B.n_estimators+1):
        _B = RandomForestClassifier(K, 
              oob_score=True, max_samples=3)
        _B.fit(x,d) 
        error.append(1-_B.oob_score_)
    return error

plt.figure()
plt.plot(oobError(B))
plt.xlabel('Liczba drzew')
plt.ylabel('''Błąd klasyfikacji na danych 
weryfikujących''')
plt.grid()
    
```

```python
# Przykład użycia permutation_importance
from numpy.random import randn
from sklearn.inspection import permutation_importance

x1=1+randn(20,5)
x2=-2+randn(20,5)
x3=3+randn(20,5)
x=np.concatenate((x1,x2,x3),axis=0)
d=np.concatenate( ([1]*20,[2]*20,[3]*20) , axis=0)


B = RandomForestClassifier(50, 
                oob_score=True, 
                max_samples=3)
B.fit(x,d) 

result = permutation_importance(B, x, d, n_repeats=50)

plt.bar(np.arange(0,5), result.importances_mean)
plt. title('''Out-of-Bag Permuted Predictor
Importance Estimates''');
plt.ylabel('Estimates');
plt.xlabel('Predictors');
```

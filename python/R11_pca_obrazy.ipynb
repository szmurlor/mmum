```python
from PIL import Image
from os import listdir,path
import numpy as np
import matplotlib.pyplot as plt

num_class = 5;  # Number of classes to read
folder_z_obrazami = (
    r'C:\Users\szmurlor\Nextcloud\Alex_obrazy20')

X = []; d  = [];
for fn in listdir(folder_z_obrazami):
    try:
        c = int(fn)
    except:
        c = -1
    if c > 0 and c <= num_class:
        for ifn in listdir(
                path.join(folder_z_obrazami, fn)):
            aimg = np.asarray(
                Image.open(
                    path.join(folder_z_obrazami, fn,ifn))
            )
            X.append(aimg.flatten())
            d.append(c)
            
X = np.array(X, dtype=np.float64)
d = np.array(d)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
X_pca = pca.transform(X)

from sklearn.manifold import TSNE
sne = TSNE(n_components=2)
X_sne = sne.fit_transform(X)

plt.subplot(1,2,1)
plt.scatter(X_pca[:,0], X_pca[:,1], c=d)
plt.title('PCA')
plt.subplot(1,2,2)
plt.title('TSNE')
plt.scatter(X_sne[:,0], X_sne[:,1], c=d)
```

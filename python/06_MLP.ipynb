```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from numpy.random import randn
x1=1+randn(20,5)
x2=-2+randn(20,5)
x3=3+randn(20,5)
x=np.concatenate((x1,x2,x3),axis=0)
d=np.concatenate( ([1]*20,[2]*20,[3]*20) , axis=0)
```

```python
from sklearn.neural_network import (
    MLPClassifier)
from sklearn.model_selection import (
    train_test_split)
(x_train, x_test, 
 d_train, d_test) = train_test_split(
                x, d,test_size=0.2)
net = MLPClassifier(hidden_layer_sizes=[5,2], 
                    activation='logistic',
                    max_iter=300, 
                    solver='lbfgs')

net.fit(x_train,d_train)

from sklearn.metrics import accuracy_score

print(f'''Acc: {accuracy_score(d_test, 
net.predict(x_test)):.3e}''')
print(f'Loss: {net.loss_:.3e}')
```

```python
x = np.arange(0,12,0.05)
x = np.array([x]).T
s = np.sin(x)
t = np.mod( x, np.pi)*2/np.pi-1
r = np.sign(np.sin(x))
c = np.abs(np.sinc(x))

plt.plot(x,s, '.r', ms=1)
plt.plot(x,t, '--c')
plt.plot(x,r, 'b')
plt.plot(x,c, '-.g')
plt.xlabel('Zmienna wejciowa x')
plt.ylabel('Wartoci zadane')
plt.title('Przebieg wartoci zadanych funkcji')
plt.ylim([-1.5,1.5])


d = np.concatenate( [s,t,r,c], axis=1 )
n = 200

from sklearn.neural_network import (
    MLPRegressor)
net = MLPRegressor([n], activation='tanh', 
                   solver='lbfgs', 
                   max_iter=120000,
                   tol=1e-10, 
                   learning_rate_init=0.1,
                   learning_rate='constant',
                   n_iter_no_change=100,
                   alpha=0.0000000001,
                   max_fun=80000)
net.fit(x,d)

x_test = np.array([ np.arange(0,12,0.053) ]).T
y_test = net.predict(x_test)

plt.figure()
plt.plot(x_test,y_test[:,0], '.r', ms=1)
plt.plot(x_test,y_test[:,1], '--c')
plt.plot(x_test,y_test[:,2], 'b')
plt.plot(x_test,y_test[:,3], '-.g')
plt.xlabel('Zmienna wejciowa x')
plt.ylabel('Wartoci zadane')
plt.ylim([-1.5,1.5])
plt.title(f'''Przebieg aproksymacji przy
 {n} neuronach''')
```
